```python
import CGREncoder
import KVRPlot
```

```python
encoder = CGREncoder.EncodingByTreatMent('all', True, 'Kallisto', 0)
encoder.multicore_convert(2, 14)
```

```python
result = encoder.getConvertedGroup()
```

```python
len(result.keys())
```

```python
result.values()
```

```python
real_value = result._getvalue()
```

```python
import pickle
pkl_out = open('CGRAllKallisto.pkl', 'wb')
pickle.dump(real_value, pkl_out)
pkl_out.close()
```

```python
import json
with open('CGRAllKallisto.json', 'w') as fp:
    json.dump(real_value, fp)
fp.close()
```
